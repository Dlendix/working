## üìå –ó–∞–¥–∞—á–∞ 1: –î–∏–Ω–∞–º–∏—á–µ—Å–∫–æ–µ –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏–µ (–†–µ–≥–∏–æ–Ω–∞–ª—å–Ω—ã–π —ç—Ç–∞–ø –í–°–û–® 2023)

**–£—Å–ª–æ–≤–∏–µ:** –î–∞–Ω–æ –¥–µ—Ä–µ–≤–æ —Å N –≤–µ—Ä—à–∏–Ω–∞–º–∏. –ö–∞–∂–¥–∞—è –≤–µ—Ä—à–∏–Ω–∞ –∏–º–µ–µ—Ç –≤–µ—Å. –ù—É–∂–Ω–æ –≤—ã–±—Ä–∞—Ç—å –º–Ω–æ–∂–µ—Å—Ç–≤–æ –≤–µ—Ä—à–∏–Ω –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ —Å—É–º–º–∞—Ä–Ω–æ–≥–æ –≤–µ—Å–∞ —Ç–∞–∫, —á—Ç–æ–±—ã –Ω–∏–∫–∞–∫–∏–µ –¥–≤–µ –≤—ã–±—Ä–∞–Ω–Ω—ã–µ –≤–µ—Ä—à–∏–Ω—ã –Ω–µ –±—ã–ª–∏ —Å–º–µ–∂–Ω—ã–º–∏.

**–†–µ—à–µ–Ω–∏–µ:** –î–ü –ø–æ –¥–µ—Ä–µ–≤—É —Å –¥–≤—É–º—è —Å–æ—Å—Ç–æ—è–Ω–∏—è–º–∏: dp[v][0] - –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –≤–µ—Å –ø–æ–¥–¥–µ—Ä–µ–≤–∞ v, –µ—Å–ª–∏ v –Ω–µ –≤—ã–±—Ä–∞–Ω–∞, dp[v][1] - –µ—Å–ª–∏ –≤—ã–±—Ä–∞–Ω–∞.

````c++
#include <bits/stdc++.h>
using namespace std;

typedef long long ll;
const int MAXN = 1e5 + 5;

vector<int> g[MAXN];
ll a[MAXN], dp[MAXN][2];
int n;

void dfs(int v, int p = -1) {
    dp[v][0] = 0;
    dp[v][1] = a[v];
    
    for (int to : g[v]) {
        if (to == p) continue;
        dfs(to, v);
        
        dp[v][0] += max(dp[to][0], dp[to][1]);
        dp[v][1] += dp[to][0];
    }
}

void solve() {
    cin >> n;
    for (int i = 1; i <= n; i++) {
        cin >> a[i];
    }
    
    for (int i = 0; i < n - 1; i++) {
        int u, v;
        cin >> u >> v;
        g[u].push_back(v);
        g[v].push_back(u);
    }
    
    dfs(1);
    cout << max(dp[1][0], dp[1][1]) << endl;
}

int main() {
    ios_base::sync_with_stdio(false);
    cin.tie(nullptr);
    solve();
    return 0;
}
````

## üìå –ó–∞–¥–∞—á–∞ 2: –î–µ—Ä–µ–≤–æ –æ—Ç—Ä–µ–∑–∫–æ–≤ —Å –æ—Ç–ª–æ–∂–µ–Ω–Ω—ã–º –æ–±–Ω–æ–≤–ª–µ–Ω–∏–µ–º (–ú—É–Ω–∏—Ü–∏–ø–∞–ª—å–Ω—ã–π —ç—Ç–∞–ø –í–°–û–® 2024)

**–£—Å–ª–æ–≤–∏–µ:** –î–∞–Ω –º–∞—Å—Å–∏–≤ –∏–∑ N —á–∏—Å–µ–ª. –í—ã–ø–æ–ª–Ω–∏—Ç—å M –æ–ø–µ—Ä–∞—Ü–∏–π –¥–≤—É—Ö —Ç–∏–ø–æ–≤:
1. –ü—Ä–∏–±–∞–≤–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ X –∫–æ –≤—Å–µ–º —ç–ª–µ–º–µ–Ω—Ç–∞–º –Ω–∞ –æ—Ç—Ä–µ–∑–∫–µ [L, R]
2. –ù–∞–π—Ç–∏ —Å—É–º–º—É —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –Ω–∞ –æ—Ç—Ä–µ–∑–∫–µ [L, R]

**–û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è:** 1 ‚â§ N, M ‚â§ 10‚Åµ, –∑–Ω–∞—á–µ–Ω–∏—è –¥–æ 10‚Åπ

**–†–µ—à–µ–Ω–∏–µ:** –î–µ—Ä–µ–≤–æ –æ—Ç—Ä–µ–∑–∫–æ–≤ —Å lazy propagation –¥–ª—è —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ–π –æ–±—Ä–∞–±–æ—Ç–∫–∏ –æ–ø–µ—Ä–∞—Ü–∏–π.

````c++
#include <bits/stdc++.h>
using namespace std;

typedef long long ll;
const int MAXN = 4e5 + 5;

ll tree[MAXN];
ll lazy[MAXN];

void push(int v, int tl, int tr) {
    if (lazy[v] != 0) {
        tree[v] += lazy[v] * (tr - tl + 1);
        if (tl != tr) {
            lazy[2 * v] += lazy[v];
            lazy[2 * v + 1] += lazy[v];
        }
        lazy[v] = 0;
    }
}

void update(int v, int tl, int tr, int l, int r, ll add) {
    push(v, tl, tr);
    if (l > tr || r < tl) return;
    if (l <= tl && tr <= r) {
        lazy[v] += add;
        push(v, tl, tr);
        return;
    }
    int tm = (tl + tr) / 2;
    update(2 * v, tl, tm, l, r, add);
    update(2 * v + 1, tm + 1, tr, l, r, add);
    tree[v] = tree[2 * v] + tree[2 * v + 1];
}

ll query(int v, int tl, int tr, int l, int r) {
    push(v, tl, tr);
    if (l > tr || r < tl) return 0;
    if (l <= tl && tr <= r) return tree[v];
    int tm = (tl + tr) / 2;
    return query(2 * v, tl, tm, l, r) + 
           query(2 * v + 1, tm + 1, tr, l, r);
}

void solve() {
    int n, m;
    cin >> n >> m;
    
    for (int i = 1; i <= n; i++) {
        ll x;
        cin >> x;
        update(1, 1, n, i, i, x);
    }
    
    while (m--) {
        int type;
        cin >> type;
        if (type == 1) {
            int l, r;
            ll x;
            cin >> l >> r >> x;
            update(1, 1, n, l, r, x);
        } else {
            int l, r;
            cin >> l >> r;
            cout << query(1, 1, n, l, r) << '\n';
        }
    }
}

int main() {
    ios_base::sync_with_stdio(false);
    cin.tie(nullptr);
    solve();
    return 0;
}
````

## üìå –ó–∞–¥–∞—á–∞ 3: –°–∏—Å—Ç–µ–º–∞ –Ω–µ–ø–µ—Ä–µ—Å–µ–∫–∞—é—â–∏—Ö—Å—è –º–Ω–æ–∂–µ—Å—Ç–≤ (DSU) (–†–µ–≥–∏–æ–Ω–∞–ª—å–Ω—ã–π —ç—Ç–∞–ø –í–°–û–® 2022)

**–£—Å–ª–æ–≤–∏–µ:** –î–∞–Ω –Ω–µ–æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –≥—Ä–∞—Ñ —Å N –≤–µ—Ä—à–∏–Ω–∞–º–∏ –∏ M —Ä–µ–±—Ä–∞–º–∏. –û–±—Ä–∞–±–æ—Ç–∞—Ç—å Q –∑–∞–ø—Ä–æ—Å–æ–≤:
1. –£–¥–∞–ª–∏—Ç—å —Ä–µ–±—Ä–æ –º–µ–∂–¥—É u –∏ v
2. –ü—Ä–æ–≤–µ—Ä–∏—Ç—å, —Å–æ–µ–¥–∏–Ω–µ–Ω—ã –ª–∏ –≤–µ—Ä—à–∏–Ω—ã u –∏ v

**–û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è:** 1 ‚â§ N, M, Q ‚â§ 2¬∑10‚Åµ

**–†–µ—à–µ–Ω–∏–µ:** –û–±—Ä–∞–±–æ—Ç–∫–∞ –∑–∞–ø—Ä–æ—Å–æ–≤ –≤ –æ–±—Ä–∞—Ç–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º DSU.

````c++
#include <bits/stdc++.h>
using namespace std;

struct DSU {
    vector<int> parent, size;
    DSU(int n) {
        parent.resize(n + 1);
        size.resize(n + 1, 1);
        for (int i = 1; i <= n; i++) parent[i] = i;
    }
    
    int find(int x) {
        if (parent[x] != x) parent[x] = find(parent[x]);
        return parent[x];
    }
    
    bool unite(int x, int y) {
        x = find(x), y = find(y);
        if (x == y) return false;
        if (size[x] < size[y]) swap(x, y);
        parent[y] = x;
        size[x] += size[y];
        return true;
    }
    
    bool connected(int x, int y) {
        return find(x) == find(y);
    }
};

void solve() {
    int n, m, q;
    cin >> n >> m >> q;
    
    map<pair<int, int>, bool> edge_exists;
    vector<tuple<int, int, int>> edges;
    
    for (int i = 0; i < m; i++) {
        int u, v;
        cin >> u >> v;
        if (u > v) swap(u, v);
        edge_exists[{u, v}] = true;
        edges.push_back({u, v, 0});
    }
    
    vector<tuple<int, int, int>> queries;
    for (int i = 0; i < q; i++) {
        int type, u, v;
        cin >> type >> u >> v;
        if (u > v) swap(u, v);
        queries.push_back({type, u, v});
        if (type == 1) {
            edge_exists[{u, v}] = false;
        }
    }
    
    DSU dsu(n);
    for (int i = 0; i < m; i++) {
        auto [u, v, _] = edges[i];
        if (edge_exists[{u, v}]) {
            dsu.unite(u, v);
        }
    }
    
    vector<string> ans;
    for (int i = q - 1; i >= 0; i--) {
        auto [type, u, v] = queries[i];
        if (type == 1) {
            dsu.unite(u, v);
        } else {
            ans.push_back(dsu.connected(u, v) ? "YES" : "NO");
        }
    }
    
    reverse(ans.begin(), ans.end());
    for (string& s : ans) cout << s << '\n';
}

int main() {
    ios_base::sync_with_stdio(false);
    cin.tie(nullptr);
    solve();
    return 0;
}
````

## üìå –ó–∞–¥–∞—á–∞ 4: –ê–ª–≥–æ—Ä–∏—Ç–º –î–µ–π–∫—Å—Ç—Ä—ã —Å –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∏–µ–º –ø—É—Ç–∏ (–ú—É–Ω–∏—Ü–∏–ø–∞–ª—å–Ω—ã–π —ç—Ç–∞–ø –í–°–û–® 2023)

**–£—Å–ª–æ–≤–∏–µ:** –î–∞–Ω –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –≤–∑–≤–µ—à–µ–Ω–Ω—ã–π –≥—Ä–∞—Ñ. –ù–∞–π—Ç–∏ –∫—Ä–∞—Ç—á–∞–π—à–∏–π –ø—É—Ç—å –∏–∑ –≤–µ—Ä—à–∏–Ω—ã 1 –≤ –≤–µ—Ä—à–∏–Ω—É N –∏ –≤—ã–≤–µ—Å—Ç–∏ –µ–≥–æ –¥–ª–∏–Ω—É –∏ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –≤–µ—Ä—à–∏–Ω.

**–û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è:** 1 ‚â§ N ‚â§ 10‚Åµ, 1 ‚â§ M ‚â§ 2¬∑10‚Åµ, –≤–µ—Å–∞ —Ä–µ–±–µ—Ä –¥–æ 10‚Åπ

**–†–µ—à–µ–Ω–∏–µ:** –ê–ª–≥–æ—Ä–∏—Ç–º –î–µ–π–∫—Å—Ç—Ä—ã —Å —Ö—Ä–∞–Ω–µ–Ω–∏–µ–º –ø—Ä–µ–¥—ã–¥—É—â–∏—Ö –≤–µ—Ä—à–∏–Ω –¥–ª—è –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∏—è –ø—É—Ç–∏.

````c++
#include <bits/stdc++.h>
using namespace std;

typedef long long ll;
typedef pair<ll, int> pli;
const ll INF = 1e18;
const int MAXN = 1e5 + 5;

vector<pair<int, ll>> g[MAXN];
ll dist[MAXN];
int prev_node[MAXN];

void solve() {
    int n, m;
    cin >> n >> m;
    
    for (int i = 0; i < m; i++) {
        int u, v;
        ll w;
        cin >> u >> v >> w;
        g[u].push_back({v, w});
    }
    
    for (int i = 1; i <= n; i++) {
        dist[i] = INF;
        prev_node[i] = -1;
    }
    
    priority_queue<pli, vector<pli>, greater<pli>> pq;
    dist[1] = 0;
    pq.push({0, 1});
    
    while (!pq.empty()) {
        auto [d, v] = pq.top();
        pq.pop();
        
        if (d != dist[v]) continue;
        
        for (auto [to, w] : g[v]) {
            if (dist[to] > dist[v] + w) {
                dist[to] = dist[v] + w;
                prev_node[to] = v;
                pq.push({dist[to], to});
            }
        }
    }
    
    if (dist[n] == INF) {
        cout << "-1" << endl;
        return;
    }
    
    cout << dist[n] << endl;
    
    vector<int> path;
    int cur = n;
    while (cur != -1) {
        path.push_back(cur);
        cur = prev_node[cur];
    }
    
    reverse(path.begin(), path.end());
    for (int v : path) {
        cout << v << " ";
    }
    cout << endl;
}

int main() {
    ios_base::sync_with_stdio(false);
    cin.tie(nullptr);
    solve();
    return 0;
}
````

## üìå –ó–∞–¥–∞—á–∞ 5: –ë–∏–Ω–∞—Ä–Ω—ã–π –ø–æ–∏—Å–∫ –ø–æ –æ—Ç–≤–µ—Ç—É (–†–µ–≥–∏–æ–Ω–∞–ª—å–Ω—ã–π —ç—Ç–∞–ø –í–°–û–® 2024)

**–£—Å–ª–æ–≤–∏–µ:** –î–∞–Ω –º–∞—Å—Å–∏–≤ –∏–∑ N —á–∏—Å–µ–ª. –ù—É–∂–Ω–æ –≤—ã–±—Ä–∞—Ç—å K –ø–æ–¥–º–∞—Å—Å–∏–≤–æ–≤ –¥–ª–∏–Ω—ã L —Ç–∞–∫, —á—Ç–æ–±—ã –º–∏–Ω–∏–º–∞–ª—å–Ω–∞—è —Å—É–º–º–∞ —Å—Ä–µ–¥–∏ –≤—ã–±—Ä–∞–Ω–Ω—ã—Ö –ø–æ–¥–º–∞—Å—Å–∏–≤–æ–≤ –±—ã–ª–∞ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π.

**–û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è:** 1 ‚â§ N ‚â§ 10‚Åµ, 1 ‚â§ K ‚â§ N, 1 ‚â§ L ‚â§ N

**–†–µ—à–µ–Ω–∏–µ:** –ë–∏–Ω–∞—Ä–Ω—ã–π –ø–æ–∏—Å–∫ –ø–æ –æ—Ç–≤–µ—Ç—É —Å –ø—Ä–æ–≤–µ—Ä–∫–æ–π –∑–∞ O(N) —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –ø—Ä–µ—Ñ–∏–∫—Å–Ω—ã—Ö —Å—É–º–º.

````c++
#include <bits/stdc++.h>
using namespace std;

typedef long long ll;
const int MAXN = 1e5 + 5;

int n, k, l;
ll a[MAXN], prefix[MAXN];

bool check(ll x) {
    // –ü—Ä–æ–≤–µ—Ä—è–µ–º, –º–æ–∂–Ω–æ –ª–∏ –≤—ã–±—Ä–∞—Ç—å k –ø–æ–¥–º–∞—Å—Å–∏–≤–æ–≤ –¥–ª–∏–Ω—ã l
    // —Å —Å—É–º–º–æ–π –Ω–µ –º–µ–Ω—å—à–µ x
    int count = 0;
    
    for (int i = l; i <= n; i++) {
        ll sum = prefix[i] - prefix[i - l];
        if (sum >= x) {
            count++;
            i += l - 1; // –ø—Ä–æ–ø—É—Å–∫–∞–µ–º —Å–ª–µ–¥—É—é—â–∏–µ l-1 —ç–ª–µ–º–µ–Ω—Ç–æ–≤
        }
    }
    
    return count >= k;
}

void solve() {
    cin >> n >> k >> l;
    
    for (int i = 1; i <= n; i++) {
        cin >> a[i];
        prefix[i] = prefix[i - 1] + a[i];
    }
    
    // –ë–∏–Ω–∞—Ä–Ω—ã–π –ø–æ–∏—Å–∫ –ø–æ –æ—Ç–≤–µ—Ç—É
    ll low = 0;
    ll high = 1e15; // –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –±–æ–ª—å—à–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
    ll ans = 0;
    
    while (low <= high) {
        ll mid = (low + high) / 2;
        if (check(mid)) {
            ans = mid;
            low = mid + 1;
        } else {
            high = mid - 1;
        }
    }
    
    cout << ans << endl;
}

int main() {
    ios_base::sync_with_stdio(false);
    cin.tie(nullptr);
    solve();
    return 0;
}
````

## üìå –ó–∞–¥–∞—á–∞ 6: –ê–ª–≥–æ—Ä–∏—Ç–º –ö–Ω—É—Ç–∞-–ú–æ—Ä—Ä–∏—Å–∞-–ü—Ä–∞—Ç—Ç–∞ (KMP) (–ú—É–Ω–∏—Ü–∏–ø–∞–ª—å–Ω—ã–π —ç—Ç–∞–ø –í–°–û–® 2022)

**–£—Å–ª–æ–≤–∏–µ:** –î–∞–Ω —Ç–µ–∫—Å—Ç T –∏ —à–∞–±–ª–æ–Ω P. –ù–∞–π—Ç–∏ –≤—Å–µ –≤—Ö–æ–∂–¥–µ–Ω–∏—è —à–∞–±–ª–æ–Ω–∞ –≤ —Ç–µ–∫—Å—Ç. –í—ã–≤–µ—Å—Ç–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—Ö–æ–∂–¥–µ–Ω–∏–π –∏ –ø–æ–∑–∏—Ü–∏–∏ –Ω–∞—á–∞–ª–∞ –∫–∞–∂–¥–æ–≥–æ –≤—Ö–æ–∂–¥–µ–Ω–∏—è (1-–∏–Ω–¥–µ–∫—Å–∞—Ü–∏—è).

**–û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è:** |T|, |P| ‚â§ 10‚Å∂

**–†–µ—à–µ–Ω–∏–µ:** –ê–ª–≥–æ—Ä–∏—Ç–º –ö–Ω—É—Ç–∞-–ú–æ—Ä—Ä–∏—Å–∞-–ü—Ä–∞—Ç—Ç–∞ –¥–ª—è –ª–∏–Ω–µ–π–Ω–æ–≥–æ –ø–æ–∏—Å–∫–∞ –ø–æ–¥—Å—Ç—Ä–æ–∫–∏.

````c++
#include <bits/stdc++.h>
using namespace std;

vector<int> compute_prefix_function(const string& s) {
    int n = s.length();
    vector<int> pi(n);
    
    for (int i = 1; i < n; i++) {
        int j = pi[i - 1];
        while (j > 0 && s[i] != s[j]) {
            j = pi[j - 1];
        }
        if (s[i] == s[j]) {
            j++;
        }
        pi[i] = j;
    }
    
    return pi;
}

void solve() {
    string text, pattern;
    cin >> text >> pattern;
    
    string s = pattern + '#' + text;
    vector<int> pi = compute_prefix_function(s);
    
    int n = text.length();
    int m = pattern.length();
    
    vector<int> positions;
    for (int i = m + 1; i < s.length(); i++) {
        if (pi[i] == m) {
            positions.push_back(i - 2 * m + 1); // 0-–∏–Ω–¥–µ–∫—Å–∞—Ü–∏—è
        }
    }
    
    cout << positions.size() << endl;
    for (int pos : positions) {
        cout << pos + 1 << " "; // 1-–∏–Ω–¥–µ–∫—Å–∞—Ü–∏—è
    }
    cout << endl;
}

int main() {
    ios_base::sync_with_stdio(false);
    cin.tie(nullptr);
    solve();
    return 0;
}
````

## üìå –ó–∞–¥–∞—á–∞ 7: –î–∏–Ω–∞–º–∏—á–µ—Å–∫–æ–µ –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏–µ –Ω–∞ –ø–æ–¥–º–Ω–æ–∂–µ—Å—Ç–≤–∞—Ö (–†–µ–≥–∏–æ–Ω–∞–ª—å–Ω—ã–π —ç—Ç–∞–ø –í–°–û–® 2023)

**–£—Å–ª–æ–≤–∏–µ:** –î–∞–Ω–æ N –ø—Ä–µ–¥–º–µ—Ç–æ–≤, –∫–∞–∂–¥—ã–π –∏–º–µ–µ—Ç –≤–µ—Å w[i] –∏ —Å—Ç–æ–∏–º–æ—Å—Ç—å c[i]. –û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ –ø–æ –≤–µ—Å—É - W. –ù–∞–π—Ç–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é —Å—Ç–æ–∏–º–æ—Å—Ç—å –ø—Ä–µ–¥–º–µ—Ç–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–∂–Ω–æ –≤–∑—è—Ç—å, –µ—Å–ª–∏ –º–æ–∂–Ω–æ –±—Ä–∞—Ç—å –Ω–µ –±–æ–ª–µ–µ –æ–¥–Ω–æ–≥–æ –ø—Ä–µ–¥–º–µ—Ç–∞ –∫–∞–∂–¥–æ–≥–æ —Ç–∏–ø–∞.

**–û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è:** 1 ‚â§ N ‚â§ 20, 1 ‚â§ W ‚â§ 10‚Åπ, w[i], c[i] ‚â§ 10‚Åπ

**–†–µ—à–µ–Ω–∏–µ:** Meet-in-the-middle –¥–ª—è –∑–∞–¥–∞—á–∏ –æ —Ä—é–∫–∑–∞–∫–µ —Å N ‚â§ 40.

````c++
#include <bits/stdc++.h>
using namespace std;

typedef long long ll;
const int MAXN = 40;

struct Item {
    ll weight, cost;
};

void solve() {
    int n;
    ll W;
    cin >> n >> W;
    
    vector<Item> items(n);
    for (int i = 0; i < n; i++) {
        cin >> items[i].weight >> items[i].cost;
    }
    
    int n1 = n / 2;
    int n2 = n - n1;
    
    vector<pair<ll, ll>> left;
    // –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º –≤—Å–µ –ø–æ–¥–º–Ω–æ–∂–µ—Å—Ç–≤–∞ –¥–ª—è –ª–µ–≤–æ–π —á–∞—Å—Ç–∏
    for (int mask = 0; mask < (1 << n1); mask++) {
        ll total_weight = 0, total_cost = 0;
        for (int i = 0; i < n1; i++) {
            if (mask & (1 << i)) {
                total_weight += items[i].weight;
                total_cost += items[i].cost;
            }
        }
        if (total_weight <= W) {
            left.push_back({total_weight, total_cost});
        }
    }
    
    // –°–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ –≤–µ—Å—É –∏ –æ—Å—Ç–∞–≤–ª—è–µ–º —Ç–æ–ª—å–∫–æ –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–µ –ø–æ —Å—Ç–æ–∏–º–æ—Å—Ç–∏
    sort(left.begin(), left.end());
    vector<pair<ll, ll>> filtered_left;
    ll max_cost = 0;
    for (auto& p : left) {
        if (p.second > max_cost) {
            max_cost = p.second;
            filtered_left.push_back(p);
        }
    }
    
    ll best = 0;
    // –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º –≤—Å–µ –ø–æ–¥–º–Ω–æ–∂–µ—Å—Ç–≤–∞ –¥–ª—è –ø—Ä–∞–≤–æ–π —á–∞—Å—Ç–∏
    for (int mask = 0; mask < (1 << n2); mask++) {
        ll total_weight = 0, total_cost = 0;
        for (int i = 0; i < n2; i++) {
            if (mask & (1 << i)) {
                total_weight += items[n1 + i].weight;
                total_cost += items[n1 + i].cost;
            }
        }
        
        if (total_weight > W) continue;
        
        // –ò—â–µ–º –ª—É—á—à–∏–π –≤–∞—Ä–∏–∞–Ω—Ç –∏–∑ –ª–µ–≤–æ–π —á–∞—Å—Ç–∏
        ll remaining = W - total_weight;
        auto it = upper_bound(filtered_left.begin(), filtered_left.end(), 
                             make_pair(remaining + 1, -1LL));
        if (it != filtered_left.begin()) {
            it--;
            best = max(best, total_cost + it->second);
        }
        best = max(best, total_cost);
    }
    
    cout << best << endl;
}

int main() {
    ios_base::sync_with_stdio(false);
    cin.tie(nullptr);
    solve();
    return 0;
}
````

## üìå –ó–∞–¥–∞—á–∞ 8: –≠—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–∞—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –∏ –æ–±—Ä–∞–±–æ—Ç–∫–∞ –∑–∞–ø—Ä–æ—Å–æ–≤ (–ú—É–Ω–∏—Ü–∏–ø–∞–ª—å–Ω—ã–π —ç—Ç–∞–ø –í–°–û–® 2024)

**–£—Å–ª–æ–≤–∏–µ:** –î–∞–Ω –º–∞—Å—Å–∏–≤ –∏–∑ N —á–∏—Å–µ–ª. –û–±—Ä–∞–±–æ—Ç–∞—Ç—å Q –∑–∞–ø—Ä–æ—Å–æ–≤:

2. –ò–∑–º–µ–Ω–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–∞ –≤ –ø–æ–∑–∏—Ü–∏–∏ pos –Ω–∞ x

**–û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è:** 1 ‚â§ N, Q ‚â§ 10‚Åµ, 1 ‚â§ K ‚â§ 100

**–†–µ—à–µ–Ω–∏–µ:** sqrt-–¥–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏—è —Å –ø—Ä–µ–¥–≤—ã—á–∏—Å–ª–µ–Ω–∏–µ–º –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –±–ª–æ–∫–∞ –∏ –∫–∞–∂–¥–æ–≥–æ –¥–µ–ª–∏—Ç–µ–ª—è.

````c++
#include <bits/stdc++.h>
using namespace std;

const int MAXN = 1e5 + 5;
const int MAXK = 101;
const int BLOCK_SIZE = 320; // sqrt(100000) ‚âà 316

int n, q;
int a[MAXN];
int block_count[(MAXN + BLOCK_SIZE - 1) / BLOCK_SIZE][MAXK];

void rebuild_block(int block_idx, int l, int r) {
    memset(block_count[block_idx], 0, sizeof(block_count[block_idx]));
    for (int i = l; i <= r; i++) {
        for (int k = 1; k < MAXK; k++) {
            if (a[i] % k == 0) {
                block_count[block_idx][k]++;
            }
        }
    }
}\nnvoid solve() {
    cin >> n;
    for (int i = 1; i <= n; i++) {
        cin >> a[i];
    }
    
    int num_blocks = (n + BLOCK_SIZE - 1) / BLOCK_SIZE;
    for (int block = 0; block < num_blocks; block++) {
        int l = block * BLOCK_SIZE + 1;
        int r = min((block + 1) * BLOCK_SIZE, n);
        rebuild_block(block, l, r);
    }
    
    cin >> q;
    while (q--) {
        int type;
        cin >> type;
        if (type == 1) {
            int l, r, k;
            cin >> l >> r >> k;
            
            if (k >= MAXK) {
                // –î–ª—è –±–æ–ª—å—à–∏—Ö K –ø—Ä–æ–≤–µ—Ä—è–µ–º –Ω–∞–ø—Ä—è–º—É—é
                int count = 0;
                for (int i = l; i <= r; i++) {
                    if (a[i] % k == 0) count++;
                }
                cout << count << endl;
                continue;
            }
            
            int count = 0;
            int block_l = (l - 1) / BLOCK_SIZE;
            int block_r = (r - 1) / BLOCK_SIZE;
            
            if (block_l == block_r) {
                // –í—Å–µ –≤ –æ–¥–Ω–æ–º –±–ª–æ–∫–µ
                for (int i = l; i <= r; i++) {
                    if (a[i] % k == 0) count++;
                }
            } else {
                // –õ–µ–≤—ã–π –Ω–µ–ø–æ–ª–Ω—ã–π –±–ª–æ–∫
                for (int i = l; i <= (block_l + 1) * BLOCK_SIZE; i++) {
                    if (i > n) break;
                    if (a[i] % k == 0) count++;
                }
                
                // –ü—Ä–∞–≤—ã–π –Ω–µ–ø–æ–ª–Ω—ã–π –±–ª–æ–∫
                for (int i = block_r * BLOCK_SIZE + 1; i <= r; i++) {
                    if (a[i] % k == 0) count++;
                }
                
                // –ü–æ–ª–Ω—ã–µ –±–ª–æ–∫–∏ –ø–æ—Å–µ—Ä–µ–¥–∏–Ω–µ
                for (int block = block_l + 1; block < block_r; block++) {
                    count += block_count[block][k];
                }
            }
            
            cout << count << endl;
        } else {
            int pos, x;
            cin >> pos >> x;
            
            int block_idx = (pos - 1) / BLOCK_SIZE;
            int l = block_idx * BLOCK_SIZE + 1;
            int r = min((block_idx + 1) * BLOCK_SIZE, n);
            
            a[pos] = x;
            rebuild_block(block_idx, l, r);
        }
    }
}

int main() {
    ios_base::sync_with_stdio(false);
    cin.tie(nullptr);
    solve();
    return 0;
}
````
